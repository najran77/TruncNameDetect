# Data preprocesing

```python
#separate names by space
def separate_string(name):

    name =[" ".join(letter.lower()) for letter in name if letter!=" "]
    return(' '.join(name))
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
```

```python
words = df['N_LST']
words
```

```python
words_ngrams = [separate_string(word) for word in words]
print((words_ngrams))
```

```python

def compute_max(w_list):
  max_mots = 0
  phrase_max_mots = ""
  # Parcourir chaque phrase dans la liste
  for phrase in w_list:
      # Diviser la phrase en mots en utilisant l'espace comme séparateur
      words = phrase.split()

      # Compter le nombre de mots dans cette phrase
      nb_mots = len(words)

      # Vérifier si cette phrase a plus de mots que la précédente maximum
      if nb_mots > max_mots:
          max_mots = nb_mots
          phrase_max_mots = phrase
  return max_mots
max_len=compute_max(words_ngrams)
max_len
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size= 26
onehot_repr = [one_hot(words, voc_size) for words in words_ngrams]
onehot_repr
```

```python
from keras.layers import Embedding
from keras.preprocessing.sequence import pad_sequences
```

```python
embedded_docs = pad_sequences(onehot_repr, padding = 'pre', maxlen = max_len+3) #all the sentences passed to embedding layer most be the same len
embedded_docs
```

```python
embedded_docs.shape
```

```python
from keras.models import Sequential
#Embedding layer
dim=50
model0 = Sequential()
model0.add(Embedding(voc_size,dim,input_length=max_len+3))

model0.compile('adam', 'mse')
```

```python
#model.summary()
```

```python
X=model0.predict(embedded_docs)
```

# Model Training

```python

from tensorflow.keras.preprocessing.text import one_hot
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Dropout
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
X_train, X_test, y_train , y_test = train_test_split(X,y, test_size=0.2, random_state=12)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
#X_train = X_train.reshape(-1,1)
X_train[10].shape
```

```python
model = Sequential()
model.add(LSTM(50 , input_shape = (20,50)))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer = 'adam')
```

```python
model.summary()
```

```python
X_train.shape
```

```python
model.fit(X_train, y_train, validation_data= (X_test, y_test), epochs = 20, batch_size = 32, verbose = 1 )
```
